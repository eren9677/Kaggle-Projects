## Column tranformation in data 

```python
ccols = [cname for cname in house_df.columns if house_df[cname].dtype =='object'and house_df[cname].nunique() <= 10  ]
ncols = [cname for cname in house_df.columns if house_df[cname].dtype in ['float64','int64']]

all_cols = ncols + ccols
house_df = house_df[all_cols]
print(house_df.columns.value_counts())
```

```python
#Importing libraries
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from snapml import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

house_df[ncols] = SimpleImputer(strategy='mean').fit_transform(house_df[ncols])
house_df[ccols] = SimpleImputer(strategy='most_frequent').fit_transform(house_df[ccols])
print(house_df.shape)
```

## Getting dummies of the dataset

```python

hot_df = pd.get_dummies(house_df[ccols])
house_df =pd.concat([house_df[ncols],hot_df],axis=1)
```

## Cleaning the Outliers from data
 

```python
from sklearn.neighbors import LocalOutlierFactor

clf = LocalOutlierFactor(contamination = 0.1)
#Predict outliers:
outliers = clf.fit_predict(house_df)
#Remove outliers where 1 represent inliers and -1 represent outliers:
train_data_cleaned = house_df[np.where(outliers == 1, True, False)]
house_df = train_data_cleaned
```

## Train test Split for models

```python

X = np.asanyarray(house_df.drop('SalePrice',axis=1))
y = np.asanyarray(house_df.SalePrice)

transform = StandardScaler()
X = transform.fit_transform(X)

x_train, x_test, y_train, y_test =train_test_split(X,y,test_size=0.35,random_state=10)
```

```python
house_df.info(verbose=True)
```

```python
tree = DecisionTreeRegressor()
tree.fit(x_train,y_train)
predictions = tree.predict(x_test)

print('mean_absolute_error of the model is : ',mean_absolute_error(predictions,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
from snapml import BoostingMachineRegressor
bmr = BoostingMachineRegressor()
bmr.fit(x_train,y_train)
bmr_predictions = bmr.predict(x_test)

print('mean_absolute_error of the model is : ',mean_absolute_error(bmr_predictions,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
predictions = rf.predict(x_test)
print('mean_absolute_error of the model is : ',mean_absolute_error(predictions,y_test))
```
